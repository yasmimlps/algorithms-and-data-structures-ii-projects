## Exercício 1: Probabilidade ao Lançar Duas Moedas

### Problema:
Duas moedas são lançadas ao mesmo tempo, cada uma carregada de forma diferente de acordo com a tabela abaixo. 
Calcule a probabilidade de todos os resultados possíveis: H-H (cara-cara), H-T (cara-coroa), T-H (coroa-cara), e T-T (coroa-coroa).

### Solução:

1. **Dados Fornecidos:**
   - As probabilidades para H (cara) para cada moeda estão listadas abaixo:
   
   | Moeda 1 (P(H)) | Moeda 2 (P(H)) |
   |----------------|----------------|
   | 0.5            | 0.5            |
   | 0.6            | 0.7            |
   | 0.4            | 0.8            |
   | 0.1            | 0.2            |
   | 0.3            | 0.4            |
   
2. **Resultados Calculados:**
   As probabilidades para cada combinação de resultados são calculadas usando as seguintes fórmulas:

   - P(H-H) = P(H1) * P(H2)
   - P(H-T) = P(H1) * (1 - P(H2))
   - P(T-H) = (1 - P(H1)) * P(H2)
   - P(T-T) = (1 - P(H1)) * (1 - P(H2))

   Usando essas fórmulas para cada par de probabilidades:

   | Resultado | P(H-H) | P(H-T) | P(T-H) | P(T-T) |
   |-----------|--------|--------|--------|--------|
   | Moeda 1: 0.5, Moeda 2: 0.5 | 0.25   | 0.25   | 0.25   | 0.25   |
   | Moeda 1: 0.6, Moeda 2: 0.7 | 0.42   | 0.18   | 0.28   | 0.12   |
   | Moeda 1: 0.4, Moeda 2: 0.8 | 0.32   | 0.08   | 0.48   | 0.12   |
   | Moeda 1: 0.1, Moeda 2: 0.2 | 0.02   | 0.08   | 0.18   | 0.72   |
   | Moeda 1: 0.3, Moeda 2: 0.4 | 0.12   | 0.18   | 0.28   | 0.42   |

3. **Interpretação dos Resultados:**
   As probabilidades para cada combinação de resultados são mostradas na tabela acima e abaixo é possível encontrar o código da solução. Podemos ver como as diferentes cargas das moedas afetam as probabilidades de cada resultado.

```python

p1_H_values = [0.5, 0.6, 0.4, 0.1, 0.3] 
p2_H_values = [0.5, 0.7, 0.8, 0.2, 0.4] 

outcomes = []

for p1_H, p2_H in zip(p1_H_values, p2_H_values):
    p_T1 = 1 - p1_H 
    p_T2 = 1 - p2_H  
    
    p_HH = p1_H * p2_H     
    p_HT = p1_H * p_T2    
    p_TH = p_T1 * p2_H     
    p_TT = p_T1 * p_T2     
    
    outcomes.append((p_HH, p_HT, p_TH, p_TT))

import pandas as pd
outcome_table = pd.DataFrame(outcomes, columns=["H-H", "H-T", "T-H", "T-T"])
outcome_table
```

## Exercício 2: Problema do Filtro de Spam com Bayes

### Problema:
60% dos e-mails na caixa de entrada são spam. Há um filtro de spam com 98% de precisão, que sinaliza e-mails de spam e não-spam. 
No final da semana, 963 e-mails estão na caixa de spam. Usando o Teorema de Bayes, calcule quantos desses e-mails na caixa de spam podem, na verdade, não ser spam.

### Solução:

1. **Dados Fornecidos:**
   - 60% dos e-mails são spam: `P(spam) = 0.60`
   - Probabilidade do filtro identificar corretamente o spam: `P(flagged_as_spam | spam) = 0.98`
   - Probabilidade do filtro identificar corretamente o não-spam: `P(not_flagged_as_spam | non_spam) = 0.98`
   - Total de e-mails sinalizados como spam: `963`

2. **Cálculo pelo Teorema de Bayes:**
   Para calcular o número de e-mails não-spam na caixa de spam:
   - Primeiro, calculamos a probabilidade de que um e-mail sinalizado seja não-spam, usando o Teorema de Bayes:
   
   \\[
   P(non\\_spam|flagged\\_as\\_spam) = \\frac{{P(flagged\\_as\\_spam|non\\_spam) \\times P(non\\_spam)}}{{P(flagged\\_as\\_spam)}}
   \\]
   
   - Onde:
     - \\(P(non\\_spam) = 0.40\\) (já que 40% dos e-mails não são spam)
     - \\(P(flagged\\_as\\_spam|non\\_spam) = 0.02\\) (probabilidade de um falso positivo)
     - \\(P(flagged\\_as\\_spam)\\) é calculado como: 
     \\[
     P(flagged\\_as\\_spam) = (P(flagged\\_as\\_spam|spam) \\times P(spam)) + (P(flagged\\_as\\_spam|non\\_spam) \\times P(non\\_spam))
     \\]
   
   Substituindo os valores, temos:
   
   \\[
   P(flagged\\_as\\_spam) = (0.98 \\times 0.60) + (0.02 \\times 0.40) = 0.592 + 0.008 = 0.60
   \\]
   
   Agora, aplicando o Teorema de Bayes:
   
   \\[
   P(non\\_spam|flagged\\_as\\_spam) = \\frac{{0.02 \\times 0.40}}{{0.60}} = 0.0133
   \\]
   
   3. **Resultado:**
   Multiplicamos a probabilidade pelo total de e-mails na caixa de spam:
   
   \\[
   0.0133 \\times 963 \\approx 13 \\text{{ e-mails na verdade não são spam}}
   \\]

Portanto, aproximadamente **13 e-mails** dos 963 sinalizados como spam na caixa de spam não são realmente spam. Abaixo, segue o código da solução. 

```python
p_spam = 0.60  # 60% dos e-mails são spam
accuracy_filter_spam = 0.98  # Precisão do filtro para detectar spam (verdadeiro positivo)
accuracy_filter_non_spam = 0.98  # Precisão do filtro para detectar não-spam (verdadeiro negativo)
total_spam_box = 963  # Total de e-mails na caixa de spam

# Probabilidade de um e-mail não ser spam (40%)
p_non_spam = 1 - p_spam

# Probabilidade de um e-mail não-spam ser incorretamente classificado como spam (falso positivo)
p_flag_non_spam = 1 - accuracy_filter_non_spam

# Probabilidade de um e-mail spam ser corretamente classificado como spam (verdadeiro positivo)
p_flag_spam = accuracy_filter_spam

# Probabilidade total de um e-mail ser classificado como spam
p_flagged_as_spam = (p_flag_spam * p_spam) + (p_flag_non_spam * p_non_spam)

# Teorema de Bayes:
p_non_spam_given_flagged_as_spam = (p_flag_non_spam * p_non_spam) / p_flagged_as_spam
non_spam_in_spam_box = p_non_spam_given_flagged_as_spam * total_spam_box

# Exibir o resultado
non_spam_in_spam_box
```

## Exercício 3: Decifração com a Cifra de César

### Problema:
Dada a string criptografada: “OCZ XJMMZXO VINRZM”, cada letra segue um processo estocástico de Markov descrito como uma Cifra de César 
com deslocamento de 7, aplicada três vezes. Isso resulta em um deslocamento total de 21 posições. A tarefa é decifrar a string.

### Solução:

1. **Entendimento da Cifra de César:**
   A Cifra de César desloca cada letra do alfabeto por um número de posições determinado. No problema, temos:
   - Deslocamento por 7 posições.
   - Aplicado três vezes: \( 7 \times 3 = 21 \) posições.

2. **Decifração:**
   Vamos aplicar um deslocamento de 21 posições para trás no alfabeto para decifrar a string original "OCZ XJMMZXO VINRZM".

3. **Resultado:**
   A string decifrada é **"THE CORRECT ANSWER"**. Abaixo segue o código pra essa solução. 

```python
# Função para decifrar uma string usando a Cifra de César com deslocamento para trás
def caesar_cipher_decrypt(text, shift):
    result = []
    for char in text:
        # Ignorar espaços
        if char == ' ':
            result.append(' ')
        # Decifrar letras maiúsculas
        elif char.isupper():
            result.append(chr((ord(char) - shift - 65) % 26 + 65))
        # Decifrar letras minúsculas
        elif char.islower():
            result.append(chr((ord(char) - shift - 97) % 26 + 97))
    return ''.join(result)

# String fornecida
encrypted_text = "OCZ XJMMZXO VINRZM"

# Aplicar a Cifra de César com deslocamento total de 21
shift = 21
decrypted_text = caesar_cipher_decrypt(encrypted_text, shift)

# Exibir o texto decifrado
decrypted_text
```

## Exercício 4: Função de Massa de Probabilidade (PMF) e Distribuição Cumulativa de Probabilidade (CDF)

### Problema:
Dadas as seguintes probabilidades para cada resultado, desenhe a **Função de Massa de Probabilidade (PMF)** e a **Distribuição Cumulativa de Probabilidade (CDF)**:

| Resultado | Probabilidade (p) |
|-----------|--------------------|
| 1         | 0.10               |
| 2         | 0.15               |
| 3         | 0.20               |
| 4         | 0.21               |
| 5         | 0.17               |
| 6         | 0.09               |
| 7         | 0.06               |
| 8         | 0.02               |

### Solução:

1. **Função de Massa de Probabilidade (PMF)**: Representa a probabilidade associada a cada resultado individual. No gráfico abaixo, a PMF para os resultados fornecidos.

2. **Distribuição Cumulativa de Probabilidade (CDF)**: Representa a soma cumulativa das probabilidades até cada resultado. O gráfico da CDF mostra a probabilidade acumulada para os resultados.

### Gráficos:
Abaixo estão os gráficos gerados para a PMF e CDF com base nas probabilidades fornecidas.

```python
import matplotlib.pyplot as plt
import numpy as np

# Dados fornecidos
outcomes = np.array([1, 2, 3, 4, 5, 6, 7, 8])
probabilities = np.array([0.10, 0.15, 0.20, 0.21, 0.17, 0.09, 0.06, 0.02])

# Função de Massa de Probabilidade (PMF)
plt.figure(figsize=(8, 6))
plt.bar(outcomes, probabilities, color='blue', alpha=0.7)
plt.title('Função de Massa de Probabilidade (PMF)')
plt.xlabel('Resultado')
plt.ylabel('Probabilidade')
plt.grid(True)
plt.show()

# Distribuição Cumulativa de Probabilidade (CDF)
cumulative_probabilities = np.cumsum(probabilities)

plt.figure(figsize=(8, 6))
plt.step(outcomes, cumulative_probabilities, where='mid', color='green', alpha=0.7)
plt.title('Distribuição Cumulativa de Probabilidade (CDF)')
plt.xlabel('Resultado')
plt.ylabel('Probabilidade Cumulativa')
plt.grid(True)
plt.show()
```

## Exercício 5: Cálculo da Entropia e Informação Mútua

### Problema:
Dado um conjunto de dados de duas variáveis categóricas (`v1` e `v2`), o objetivo é:
1. Calcular a **entropia** de cada variável, que nos diz quantos bits são necessários para codificar cada uma delas.
2. Calcular a **informação mútua** entre `v1` e `v2`, que nos diz quantos bits são "salvos" ao codificar `v1` sabendo os valores de `v2`.

### Passos:
1. Carregar os dados categóricos para `v1` e `v2`.
2. Calcular a probabilidade de cada possível valor para `v1` e `v2`.
3. Usar a função de entropia para calcular a quantidade de bits necessária para codificar cada variável.
4. Calcular a informação mútua entre `v1` e `v2` para determinar o quanto saber `v2` reduz a incerteza de `v1`.

### Resultados:
Abaixo estão as entropias de `v1` e `v2`, e o percentual de bits salvos ao saber `v2` para codificar `v1`.

```python
import pandas as pd
from scipy.stats import entropy
from sklearn.metrics import mutual_info_score

# Dados corrigidos para garantir que v1 e v2 tenham o mesmo comprimento
data = {
    'v1': ['a', 'a', 'a', 'a', 'a', 'a', 'a', 'a', 'b', 'b', 'b', 'b', 'b', 'b', 'c', 'c', 'c', 'c', 'c', 'c', 'c', 
           'd', 'd', 'd', 'd', 'd', 'd', 'd', 'd', 'e', 'e', 'e', 'e', 'e', 'e', 'e', 'e', 'e', 'e', 'e', 'f', 'f', 'f', 'f', 'f'],
    'v2': ['a', 's', 'a', 'd', 'd', 'f', 'f', 'f', 'f', 'g', 'g', 'g', 'g', 'h', 'h', 'h', 'h', 'j', 'j', 'j', 'k', 
           'j', 'g', 'h', 'h', 'g', 'f', 'f', 'f', 'f', 'd', 'd', 's', 's', 's', 's', 's', 'a', 'a', 'a', 'a', 'a', 'd', 'f', 'g']
}

# Criando o DataFrame
df = pd.DataFrame(data)

# Calculando a entropia de v1 e v2
v1_ps = (df.groupby(by="v1").size() / df.shape[0]).values
v2_ps = (df.groupby(by="v2").size() / df.shape[0]).values

v1_info = entropy(v1_ps)
v2_info = entropy(v2_ps)

# Exibindo as entropias
print(f"Entropia de v1: {v1_info}")
print(f"Entropia de v2: {v2_info}")

# Calculando a informação mútua
mi = mutual_info_score(df["v1"], df["v2"])

# Percentual de bits salvos
percent_bits_saved = (100 * mi / v1_info)
print(f"Knowing v2 gives us {percent_bits_saved:.2f}% of v1")

# Resultado final: 
v1_info, v2_info, mi, percent_bits_saved
```

### Descrição do Problema:
Mr. A considera Ms. B um amigo, mas ela não gosta dele de volta. Ela tem uma amizade recíproca com C e D, mas apenas C considera D um amigo. 
D também enviou pedidos de amizade para E, F, G, e H, mas, até agora, apenas G respondeu. G também tem uma amizade recíproca com A.

### Explicação do Grafo:
- **A -> B**: A considera B amigo, mas não há reciprocidade.
- **B -> C** e **B -> D**: Ms. B tem amizade recíproca com C e D.
- **C -> D**: Apenas C considera D amigo.
- **D -> E**, **D -> F**, **D -> G**, **D -> H**: D enviou pedidos de amizade para E, F, G e H.

## Exercício 3: Grafo Não-Direcionado e Ponderado das Amizades
### Descrição:
A conexão é ponderada com **2** se for recíproca, e **1** caso contrário.

### Explicação das Conexões:
- **A - B**: Peso 1 (não recíproco)
- **B - C**: Peso 2 (recíproco)
- **B - D**: Peso 2 (recíproco)
- **C - D**: Peso 1 (não recíproco)
- **D - E**, **D - F**, **D - G**, **D - H**: Peso 1 (não recíproco)
- **A - G**: Peso 2 (recíproco)

O grafo não-direcionado e ponderado abaixo mostra as conexões de amizade, com os pesos indicando a reciprocidade das amizades.


```python
# Criando o grafo não-direcionado e ponderado
G_weighted = nx.Graph()

# Adicionando as arestas com pesos
# Peso 2 para conexões recíprocas e 1 para as que não são
weighted_edges = [
    ('A', 'B', 1),  # Não recíproco
    ('B', 'C', 2),  # Recíproco
    ('B', 'D', 2),  # Recíproco
    ('C', 'D', 1),  # Não recíproco
    ('D', 'E', 1),  # Não recíproco
    ('D', 'F', 1),  # Não recíproco
    ('D', 'G', 2),  # Recíproco
    ('D', 'H', 1),  # Não recíproco
    ('A', 'G', 2)   # Recíproco
]

# Adicionando as arestas ponderadas ao grafo
G_weighted.add_weighted_edges_from(weighted_edges)

# Desenhando o grafo não-direcionado e ponderado
plt.figure(figsize=(8, 6))
pos = nx.spring_layout(G_weighted)  # Posição dos nós
nx.draw(G_weighted, pos, with_labels=True, node_color='lightgreen', node_size=3000, font_size=10, font_weight='bold')

# Adicionando os pesos das arestas
labels = nx.get_edge_attributes(G_weighted, 'weight')
nx.draw_networkx_edge_labels(G_weighted, pos, edge_labels=labels)

plt.title("Grafo Não-Direcionado e Ponderado das Amizades")
plt.show()
```

Exercício 1 do Capítulo 4: Grafo Bipartido
### Problema:
A rede fornecida é bipartida. O objetivo é:
1. Identificar os nós em cada conjunto (tipo).
2. Encontrar os nós com o maior número de vizinhos em cada conjunto.

### Solução:

1. **Identificação dos conjuntos bipartidos**:
   Em um grafo bipartido, os nós podem ser divididos em dois conjuntos disjuntos, onde as arestas conectam vértices de conjuntos diferentes. Com base nos pares fornecidos, podemos dividir os nós em dois conjuntos.

2. **Encontrar o nó com mais vizinhos**:
   Vamos calcular o número de vizinhos para cada nó e identificar o nó com o maior número de conexões em cada conjunto.

### Resultados:
Abaixo estão os resultados para o nó com mais vizinhos em cada conjunto.

```python
import networkx as nx

# Lista de pares fornecida no exercício
edges = [
    (1, 2), (1, 3), (4, 2), (4, 3), (4, 5), (6, 2), (7, 2), (7, 5), (8, 9), (10, 11),
    # (continuação dos pares)
    (139, 3), (140, 2), (141, 2), (141, 5)
]

# Criando o grafo bipartido
G = nx.Graph()
G.add_edges_from(edges)

# Função para identificar o maior número de vizinhos por nó
def find_most_neighbors(graph):
    # Encontrar o grau de cada nó (número de vizinhos)
    degrees = dict(graph.degree())
    # Identificar o nó com o maior número de vizinhos
    max_node = max(degrees, key=degrees.get)
    return max_node, degrees[max_node]

# Encontrando os nós com mais vizinhos em cada conjunto
# Conjuntos bipartidos serão divididos por tipo
nodes_type1 = set()
nodes_type2 = set()

for u, v in edges:
    nodes_type1.add(u)
    nodes_type2.add(v)

# Encontrar os nós com mais vizinhos em cada tipo
most_neighbors_type1, degree_type1 = find_most_neighbors(G.subgraph(nodes_type1))
most_neighbors_type2, degree_type2 = find_most_neighbors(G.subgraph(nodes_type2))

# Exibir os resultados formatados
print(f"O nó no primeiro conjunto com mais vizinhos é {most_neighbors_type1} com {degree_type1} vizinhos.")
print(f"O nó no segundo conjunto com mais vizinhos é {most_neighbors_type2} com {degree_type2} vizinhos.")
```

## Exercício 2: Rede Multilayer

### Problema:
A rede fornecida é uma rede multilayer. O objetivo é determinar se a rede tem um acoplamento do tipo estrela, clique ou cadeia.

1. **Star coupling (acoplamento em estrela)**: Um nó central conectado a vários outros, sem que esses nós estejam diretamente conectados entre si.
2. **Clique coupling (acoplamento em clique)**: Todos os nós em um conjunto estão diretamente conectados entre si.
3. **Chain coupling (acoplamento em cadeia)**: Os nós estão conectados sequencialmente.

### Solução:
A rede será analisada para identificar o padrão de conexões inter-camadas (coupling) e dentro das camadas.

### Resultados:
Abaixo está o grafo da rede multilayer com as conexões inter-camadas e a identificação dos graus de cada nó para determinar o tipo de acoplamento.

```python
import networkx as nx
import matplotlib.pyplot as plt

# Dados fornecidos
edges = [
    (1, 2, 'C'), (1, 3, 'C'), (1, 4, 'C'),
    (5, 6, 'C'), (5, 7, 'C'), (5, 8, 'C'),
    (9, 10, 'C'), (9, 11, 'C'), (9, 12, 'C'),
    (13, 14, 'C'), (13, 15, 'C'), (13, 16, 'C'),
    (1, 5, 1), (1, 9, 1), (1, 13, 1),
    (5, 9, 1), (5, 13, 1), (9, 13, 1),
    (2, 6, 2), (6, 10, 2), (10, 14, 2), (14, 2, 2),
    (3, 7, 3), (7, 11, 3), (11, 15, 3),
    (4, 8, 4), (12, 16, 4)
]

# Criando o grafo multilayer
G = nx.MultiGraph()

# Adicionando as arestas ao grafo
for u, v, layer in edges:
    G.add_edge(u, v, layer=layer)

# Verificando as conexões inter-camadas
interlayer_edges = [(u, v) for u, v, layer in G.edges(data='layer') if layer == 'C']

# Exibindo as conexões inter-camadas
print("Conexões inter-camadas (Coupling):", interlayer_edges)

# Plotando o grafo para análise visual
plt.figure(figsize=(8, 6))
pos = nx.spring_layout(G)
nx.draw(G, pos, with_labels=True, node_color='lightblue', node_size=3000, font_size=10, font_weight='bold')
plt.title("Rede Multilayer com Conexões Inter-Camadas")
plt.show()

# Verificar se a rede inter-camada segue um padrão de estrela, clique ou cadeia
degree_dict = G.degree()
print("Graus dos nós na rede:", degree_dict)
```

## Exercício 3 do Capítulo 4: Hipergrafo

### Problema:
1. Transformar o hipergrafo fornecido em uma **rede unipartida**, onde cada hiperarco é dividido em arestas conectando todos os nós pertencentes ao hiperarco.
2. Transformar o hipergrafo em uma **rede bipartida**, onde cada hiperarco é tratado como um nó de um conjunto, e os nós que participam desse hiperarco se conectam a ele.

### Solução:
Abaixo estão os resultados com as redes unipartida e bipartida desenhadas, de acordo com as regras fornecidas.

```python
import networkx as nx
import matplotlib.pyplot as plt

# Dados fornecidos (hipergrafo)
hyperedges = [
    [1, 2, 3, 4, 5, 6],
    [2, 4, 7, 8],
    [4, 5, 6, 7, 10],
    [6, 9, 10, 11, 12],
    [11, 13, 14, 15],
    [11, 12, 13],
    [14, 15, 16, 17, 18],
    [15, 17, 19, 20],
    [21, 22, 23],
    [23, 24],
    [24, 25, 26],
    [25, 27, 28, 29]
]

# Transformação para rede unipartida
G_unipartite = nx.Graph()

# Para cada hiperarco, conectar todos os nós entre si
for hyperedge in hyperedges:
    for i in range(len(hyperedge)):
        for j in range(i+1, len(hyperedge)):
            G_unipartite.add_edge(hyperedge[i], hyperedge[j])

# Desenhando a rede unipartida
plt.figure(figsize=(8, 6))
pos = nx.spring_layout(G_unipartite)
nx.draw(G_unipartite, pos, with_labels=True, node_color='lightblue', node_size=3000, font_size=10, font_weight='bold')
plt.title("Rede Unipartida")
plt.show()

# Transformação para rede bipartida
G_bipartite = nx.Graph()

# Adicionar os nós das hiperarcos
hyperedge_nodes = ['E' + str(i) for i in range(1, len(hyperedges)+1)]

# Conectar os nós aos hiperarcos
for i, hyperedge in enumerate(hyperedges):
    for node in hyperedge:
        G_bipartite.add_edge(hyperedge_nodes[i], node)

# Desenhando a rede bipartida
plt.figure(figsize=(8, 6))
pos = nx.spring_layout(G_bipartite)
nx.draw(G_bipartite, pos, with_labels=True, node_color='lightgreen', node_size=3000, font_size=10, font_weight='bold')
plt.title("Rede Bipartida")
plt.show()
```

## Exercício 4: Rede Dinâmica

### Problema:
A rede fornecida é dinâmica, com a terceira e quarta colunas indicando o primeiro e último snapshot em que a aresta esteve presente. 
Nosso objetivo é agregar essa rede usando janelas disjuntas de tamanho 3, criando redes agregadas em cada janela.

### Solução:
Abaixo estão os resultados das redes agregadas para cada janela de tempo.

```python
import networkx as nx
import matplotlib.pyplot as plt
from collections import defaultdict

# Lista de arestas fornecida
edges = [
    (1, 2, 1, 4), (1, 2, 7, 9), (1, 3, 3, 7), (1, 4, 2, 4), (1, 5, 1, 9),
    (2, 3, 1, 2), (2, 4, 1, 3), (2, 4, 6, 7), (2, 4, 9, 9), (3, 5, 5, 7),
    (3, 6, 1, 3), (3, 6, 8, 9), (4, 7, 1, 9), (5, 8, 1, 2), (5, 8, 4, 7),
    (5, 8, 9, 9), (5, 9, 1, 5), (5, 10, 8, 9), (5, 11, 7, 8), (6, 12, 4, 7),
    (6, 13, 2, 3), (7, 12, 2, 6)
]

# Definindo o tamanho da janela
window_size = 3

# Função para agregar a rede dinâmica em janelas disjuntas
def aggregate_by_window(edges, window_size):
    # Dicionário para armazenar arestas por janela
    aggregated_edges = defaultdict(list)
    
    # Iterar sobre as arestas
    for u, v, start, end in edges:
        # Dividir o tempo em janelas disjuntas de tamanho window_size
        for window_start in range(1, end + 1, window_size):
            window_end = window_start + window_size - 1
            
            # Verificar se a aresta está ativa nessa janela
            if start <= window_end and end >= window_start:
                aggregated_edges[(window_start, window_end)].append((u, v))
    
    return aggregated_edges

# Agregar a rede por janelas
aggregated_networks = aggregate_by_window(edges, window_size)

# Plotar cada janela de tempo como uma rede separada
for window, window_edges in aggregated_networks.items():
    G = nx.Graph()
    G.add_edges_from(window_edges)
    
    # Desenhar a rede agregada para a janela
    plt.figure(figsize=(8, 6))
    pos = nx.spring_layout(G)
    nx.draw(G, pos, with_labels=True, node_color='lightblue', node_size=3000, font_size=10, font_weight='bold')
    plt.title(f"Rede Agregada: Janela {window}")
    plt.show()
```
